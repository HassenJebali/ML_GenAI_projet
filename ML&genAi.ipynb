## Conclusion sur l‚Äô√©valuation du mod√®le de r√©gression

Les m√©triques obtenues lors de l‚Äô√©valuation du mod√®le montrent que les performances sont insuffisantes :

- Le **R¬≤ n√©gatif (-0.20)** indique que le mod√®le est moins performant qu‚Äôune simple moyenne, ce qui signifie qu‚Äôil n‚Äôexplique pas la variance des donn√©es.
- Les erreurs moyennes (MSE, RMSE, MAE) restent relativement √©lev√©es, sugg√©rant un √©cart important entre les pr√©dictions et les valeurs r√©elles.

Cela implique que le mod√®le actuel ne parvient pas √† capter correctement la relation entre les variables explicatives et la variable cible. Il est n√©cessaire d‚Äôenvisager :

- Une meilleure pr√©paration des donn√©es,
