## Conclusion sur l’évaluation du modèle de régression

Les métriques obtenues lors de l’évaluation du modèle montrent que les performances sont insuffisantes :

- Le **R² négatif (-0.20)** indique que le modèle est moins performant qu’une simple moyenne, ce qui signifie qu’il n’explique pas la variance des données.
- Les erreurs moyennes (MSE, RMSE, MAE) restent relativement élevées, suggérant un écart important entre les prédictions et les valeurs réelles.

Cela implique que le modèle actuel ne parvient pas à capter correctement la relation entre les variables explicatives et la variable cible. Il est nécessaire d’envisager :

- Une meilleure préparation des données,
- L’essai de modèles alternatifs plus adaptés,
- Une optimisation des hyperparamètres,
- Et potentiellement l’enrichissement du jeu de données.
